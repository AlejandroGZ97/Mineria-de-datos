```python
datos['Current_Price'] = datos['Current_Price'].str.replace('₹', '')
datos['Current_Price'] = datos['Current_Price'].str.replace(',', '')
datos['Current_Price'].unique()
datos.dropna(axis = 0, inplace = True)
datos['Current_Price'] = datos['Current_Price'].astype(int)
```

```python
datos['How_Many_Sold'] = datos['How_Many_Sold'].str.replace(',', '')
datos['How_Many_Sold'].unique()
datos.dropna(axis = 0, inplace = True)
datos['How_Many_Sold'] = datos['How_Many_Sold'].astype(int)
```

```python
plt.figure(figsize = (20,10))
sns.barplot(x = 'Brand_Name',
            y = 'How_Many_Sold',            
            data = datos).set(title = "How_Many_Sold vs Brand Name");
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize = (20,10))
sns.barplot(x = 'Brand_Name',
            y = 'RATING',            
            data = datos).set(title = "Rating vs Brand Name");
plt.xticks(rotation = 90)
plt.show()
```
