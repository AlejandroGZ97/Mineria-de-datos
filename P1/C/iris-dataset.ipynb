```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
datos = pd.read_csv('../iris.csv')
```

```python
datos.head()
```

```python
datos.describe()
```

```python
datos['species'].hist()
```

```python
#scraterplot
colors =['blue','red','green']
species = datos['species'].unique()
species

```

```python
cols = datos.columns.drop('species')
cols
```

```python
for i in range(3):
    x = datos[datos['species'] == species[i]]
    plt.scatter(x['sepal_length'], x['sepal_width'], c = colors[i],label=species[i])
    plt.xlabel("sepal_length")
    plt.ylabel("sepal_width")
```

```python
fig,ax = plt.subplots(4,4,figsize=(12,6))
for i,col_i in enumerate(cols):
    for j,col_j in enumerate(cols):
        for k in range(3):
            x = datos[datos['species'] == species[k]]
            if j==0:
                ax[i][j].set_ylabel(col_i)
            if i== len(cols)-1:
                ax[i][j].set_xlabel(col_j)
            if i == j:
                ax[i][j].hist(datos[col_i])
            else:
                ax[i][j].scatter(x[col_i],x[col_j], c = colors[k],label=species[k])
plt.tight_layout()
plt.show()
        
```
