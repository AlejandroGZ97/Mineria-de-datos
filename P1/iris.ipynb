```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('iris.csv')
```

```python
data.sample(10)
```

```python
data.describe()
```

```python
set(data.species)
```

```python
colors = {"setosa": "Blue", "versicolor": "Red", "virginica": "Green"}
species_color = data.species.map(colors)
```

```python
fig, ax = plt.subplots()
for species in set(data.species):
    ax.scatter(
        data.sepal_length[data.species == species],
        data.sepal_width[data.species == species],
        s = 30,
        c = colors[species],
        label = species)
plt.legend()
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")
plt.show()
```
